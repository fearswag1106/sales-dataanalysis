```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.histplot(data['Rating'].dropna(), kde=True, bins=20, color='Red')
plt.title('Application Ratings')
plt.xlabel('Rating')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating', data=data, color='skyblue')
plt.title('boxplot Application Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
data['Rating']=np.log1p(data['Rating'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating', data=data, color='skyblue')
plt.title('boxplot Application Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
```

```python
filterdata=data[(data['Rating']>=lower_bound)&(data['Rating']<=upper_bound)]
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating', data=filterdata, color='yellow')
plt.title('boxplot Application Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.countplot(y='Category', data=data, hue='Category', order=data['Category'].value_counts().index, palette='viridis', legend=False)
plt.title('count apps by Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```
